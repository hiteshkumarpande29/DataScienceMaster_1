Q-2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that displays values for typically two variables (or dimensions) using Cartesian coordinates. Each data point is represented as a dot (or other markers) at the intersection of its values on the x-axis (horizontal) and y-axis (vertical). Scatter plots are useful for showing the relationship or correlation between two variables.
Key Features of Scatter Plots:

Data Points: Each point represents one observation in the dataset, with its position determined by the values of two variables.
    
Axes: The x-axis represents one variable, while the y-axis represents another.
    
Relationships: Scatter plots can help show trends, clusters, or correlations between variables. For example:
        
  Positive correlation: As x increases, y increases.
        
  Negative correlation: As x increases, y decreases.
  
  No correlation: No obvious pattern between x and y.
    
  Outliers: Scatter plots can also reveal outliers—points that do not follow the general trend.

Example of a Scatter Plot:

Let's say you want to visualize the relationship between the height and weight of individuals. Each dot on the scatter plot represents one person's height (on the x-axis) and weight (on the y-axis).
Example in Python using Matplotlib:



```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows you to divide the figure into a grid and place plots within the grid. This is particularly useful when you want to display multiple plots side by side or in a specific arrangement.


Here's a code that uses the subplot() function to draw four line plots:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])   # Line 1
y2 = np.array([50, 20, 40, 20, 60, 70])       # Line 2
y3 = np.array([10, 20, 30, 40, 50, 60])       # Line 3
y4 = np.array([200, 350, 250, 550, 450, 150]) # Line 4

# Create subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o')
plt.title('Line 1: y = [0, 100, 200, 300, 400, 500]')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='o', color='r')
plt.title('Line 2: y = [50, 20, 40, 20, 60, 70]')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='o', color='g')
plt.title('Line 3: y = [10, 20, 30, 40, 50, 60]')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='o', color='m')
plt.title('Line 4: y = [200, 350, 250, 550, 450, 150]')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of chart that presents categorical data with rectangular bars. Each bar's length or height represents the value of the data. Bar plots are commonly used to compare different categories or groups.
Why is it Used?


**Comparison:** It helps in comparing the magnitude of different categories.
Clarity: It's easy to see which category has the highest or lowest value.
Categorical Data Visualization: Bar plots are suitable for visualizing data that is divided into discrete categories.

Plotting a Bar Plot and a Horizontal Bar Plot using the given data:



```python
import matplotlib.pyplot as plt
import numpy as np

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure
plt.figure(figsize=(10, 5))

# Bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit, color='blue')
plt.title('Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='green')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)


A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on five summary statistics:

Minimum – the smallest data point, excluding outliers.

First quartile (Q1) – the 25th percentile of the data.

Median (Q2) – the middle of the data set, i.e., the 50th percentile.

Third quartile (Q3) – the 75th percentile of the data.

Maximum – the largest data point, excluding outliers.

In addition, a box plot often shows outliers as individual points that fall outside the "whiskers" of the plot.

Why is a box plot used?

Visualizing Distribution: Box plots are great for showing the spread and skewness of data in a visual format.
    
Comparing Distributions: Multiple box plots can be used to compare different datasets.
    
Identifying Outliers: Any data points that are significantly higher or lower than the majority of the data are shown as outliers.

I'll plot the box plot for the provided data using the box1 and box2 arrays:




```python
import numpy as np
import matplotlib.pyplot as plt

# Generating the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Plotting the box plot
data = [box1, box2]

plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title("Box Plot of Box 1 and Box 2")
plt.ylabel("Values")
plt.show()
```

Here is the box plot for the data from box1 and box2. The plot shows the distribution of values in both datasets, including the median, quartiles, and any potential outliers.

Box 1 has a median centered around 100 with a narrower spread (since it has a lower standard deviation).
    
Box 2 has a median lower than Box 1 (around 90) and a wider spread due to its higher standard deviation. ​
